```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt
```

```python
path='morePeople.jpg'
```

```python
faceClassifier=cv2.CascadeClassifier('/var/local/anaconda3/lib/python3.6/site-packages/cv2/data/haarcascade_frontalface_default.xml')
img=Image.open(path)
plt.figure("test")
plt.imshow(img)
plt.show()
```

```python
objImage=cv2.imread(path)
```

```python
cvtImage=cv2.cvtColor(objImage,cv2.COLOR_BGR2GRAY)
```

```python
foundFaces=faceClassifier.detectMultiScale(cvtImage,scaleFactor=1.1,minNeighbors=5,minSize=(5,5))
```

```python
for (x,y,w,h) in foundFaces:
    cv2.rectangle(objImage,(x,y),(x+w,y+h),(0,0,255),2)
```

```python
result='test_finish.jpg'
```

```python
cv2.imwrite(result, objImage)
```


img=Image.open(result)


```python
plt.figure("test")
plt.imshow(img)
plt.show()
```
